```python
import pandas as pd
import numpy as np
df = pd.read_csv("final-employee-ds.csv")


```

Question : 1


Load the TrainingHours column into a NumPy array.
Convert to float datatype


Find mean & standard deviation

Final output: Two numeric values






```python

t_hours = df['TrainingHours'].to_numpy(dtype=float)
print(t_hours)
mean_v=t_hours.mean()
print('Mean value is : -> ', mean_v)
stnd_dv=t_hours.std()
print('Standard Deviation is : ->', stnd_dv)
```

Question : 2


Handle missing values (if any):
Fill missing Salary with department-wise median


Calculate total Salary sum for employees with ExperienceYears > 12 and ProjectCount > 5

Final output: One numeric value





```python
df['Salary'].fillna(df.groupby('Department')['Salary'].transform('median'))

salary_sum = df.loc[(df['ExperienceYears'] > 12) & (df['ProjectCount'] > 5), 'Salary'].sum()
print(salary_sum)

```

Question : 3



Filter employees who:
Department = "IT" or "Finance"


Age between 30–45


PerformanceScore > 88


Sort by Salary descending

Final output: Table with Name, Department, Salary, PerformanceScore





```python
filtered_df = df.loc[(df['Department'].str.contains('IT|Finance', na=False) ) & ((df['Age'] > 30) & (df['Age'] <45)) & (df['PerformanceScore'] > 88)]
df
filtered_df

sorted_df = filtered_df.sort_values('Salary', ascending=False)
output=sorted_df[['Name', 'Department', 'Salary', 'PerformanceScore']]
print(output)
```

Question : 4


Add new column SalaryPerHour = Salary ÷ (WorkHoursPerWeek × 4)
Filter top 5 employees by SalaryPerHour

Final output: Table(5 Rows) with Name, SalaryPerHour







```python
df['SalaryPerHour'] = df['Salary'] / (df['WorkHoursPerWeek'] * 4)
df
df1 = df.sort_values('SalaryPerHour', ascending=False)
df1
df1=df1[['Name', 'SalaryPerHour']].head(5).sort_values('SalaryPerHour', ascending=False)
df1
```
